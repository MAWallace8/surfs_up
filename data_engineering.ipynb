```python
#import dependencies

import csv
import pandas as pd
```

```python
#call variables to provided csv files

measurements = "hawaii_measurements.csv"
stations = "hawaii_stations.csv"
```

```python
#convert stations to dataframe from csv

stations_pd = pd.read_csv(stations)
stations_pd.head(-1)
```

```python
#convert measurements to dataframe from csv

measurements_pd = pd.read_csv(measurements)
measurements_pd.head(-1)
```

```python
#outer merge of both dataframes on station

stations_measurements_pd = pd.merge(measurements_pd, stations_pd, on='station', how='outer')
stations_measurements_pd.head(-1)
```

```python
#fill NaNs with 0, NaN most likely 0 precipitation for that day

clean_stations_measurements_pd = stations_measurements_pd.fillna(0)
clean_stations_measurements_pd.head(-1)
```

```python
#export cleaned data to csv

clean_stations_measurements_pd.to_csv('clean_weather.csv')
```
